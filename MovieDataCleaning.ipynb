```python
import pandas as pd
import numpy as np

pd.options.display.max_columns = 30
```

For this project, we will use the movies data set from kaggle, since the movie database don't support a querry to get all data.
<br>
This data set can be download from: [The Movies Dataset](https://www.kaggle.com/datasets/rounakbanik/the-movies-dataset)
<br>
This dataset consists of the following files:<br>

movies_metadata.csv: The main Movies Metadata file. Contains information on 45,000 movies featured in the Full MovieLens dataset. Features include posters, backdrops, budget, revenue, release dates, languages, production countries and companies.<br>

keywords.csv: Contains the movie plot keywords for our MovieLens movies. Available in the form of a stringified JSON Object.<br>

credits.csv: Consists of Cast and Crew Information for all our movies. Available in the form of a stringified JSON Object.<br>

links.csv: The file that contains the TMDB and IMDB IDs of all the movies featured in the Full MovieLens dataset.<br>

links_small.csv: Contains the TMDB and IMDB IDs of a small subset of 9,000 movies of the Full Dataset.<br>

ratings_small.csv: The subset of 100,000 ratings from 700 users on 9,000 movies.<br>

The Full MovieLens Dataset consisting of 26 million ratings and 750,000 tag applications from 270,000 users on all the 45,000 movies in this dataset can be accessed here

```python
# load data
df = pd.read_csv('data/movies_metadata.csv', low_memory=False)
```

```python
df.head()
```

```python
# Dropping irrelevant columns
df.drop(columns=['adult', 'imdb_id', 'original_title', 'video', 'homepage'], inplace=True)
```

```python
df.info()
```

#### Handling JSON columns

```python
import json
import ast
```

```python
df.genres[0]
```

```python
# remove the quotation mark at the beginning for each row of the column genres
df.genres = df.genres.apply(ast.literal_eval)
```

```python
df.belongs_to_collection
```

```python
df.belongs_to_collection[0]
```

```python
# Remove the quotation mark for relevant values, and set nan value to np.nan
df.belongs_to_collection = df.belongs_to_collection.apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else np.nan)
```

```python
# similar to 'spoken_languages', 'production_countries' and 'production_companies' 
df.spoken_languages = df.spoken_languages.apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else np.nan)
df.production_countries = df.production_countries.apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else np.nan)
df.production_companies = df.production_companies.apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else np.nan)
```

```python
df.head()
```

#### Flatten nested columns

```python
# example of nested value
df.genres[0]
```

```python
# for genres column, we just need the names of the genres
df.genres = df.genres.apply(lambda x: '|'.join(i['name'] for i in x))
```

```python
df.genres[0]
```

```python
df.genres.value_counts(dropna=False).head()
```

```python
# set the empty value to np.nan
df.genres.replace('', np.nan, inplace=True)
```

```python
# Similar to other nested columns
df.spoken_languages = df.spoken_languages.apply(lambda x: '|'.join(i['name'] for i in x) if isinstance(x, list) else np.nan)
df.spoken_languages.replace('', np.nan, inplace=True)

df.production_countries = df.production_countries.apply(lambda x: '|'.join(i['name'] for i in x) if isinstance(x, list) else np.nan)
df.production_countries.replace('', np.nan, inplace=True)

df.production_companies = df.production_companies.apply(lambda x: '|'.join(i['name'] for i in x) if isinstance(x, list) else np.nan)
df.production_companies.replace('', np.nan, inplace=True)

df.belongs_to_collection = df.belongs_to_collection.apply(lambda x: x['name'] if isinstance(x, dict) else np.nan) 
```

#### Cleaning Numerical Columns

```python
df.head(2)
```

```python
df.info()
```

```python
# conver budget columns to float
df.budget = pd.to_numeric(df.budget, errors='coerce')
```

```python
df.budget.value_counts(dropna=False)
```

```python
# change 0 values to np.nan and display the budget as million usd
df.budget = df.budget.replace(0, np.nan)
df.budget = df.budget.div(1000000)
```

```python
# same thing to revenue columns
df.revenue = df.revenue.replace(0, np.nan)
df.revenue = df.revenue.div(1000000)
```

```python
# rename columns
df.rename(columns = {"revenue":"revenue_musd", "budget":"budget_musd"}, inplace = True)
```

```python
df.runtime.value_counts(dropna=False).head(10)
```

```python
# conver 0 to np.nan
df.runtime = df.runtime.replace(0, np.nan)
```

```python
# convert ID, popularity to numeric
df.id = pd.to_numeric(df.id, errors='coerce')
df.popularity = pd.to_numeric(df.popularity, errors='coerce')
```

```python
# set np.nan to vote_average value where vote_Count =0
df.loc[df.vote_count == 0, 'vote_average'] = np.nan
```

#### Cleaning datatime columns

```python
df.info()
```

```python
df.release_date = pd.to_datetime(df.release_date, errors = 'coerce')
```

#### Cleaning Text / String Columns

#### Clean the data one more time :v

```python
df.cast = df.cast.apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else np.nan)
```

```python
df.crew = df.crew.apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else np.nan)
```

```python
df['crew_size'] = df.crew.apply(lambda x: len(x))
```

```python
# get the director for each movie
def get_director(x):
    for i in x:
        if i['job'] == 'Director':
            return i['name']
    return np.nan
    
```

```python
df['director'] = df.crew.apply(get_director)
```
